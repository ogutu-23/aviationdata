## Import Pandas
#### Create a new dataframe and name it 

```python
import pandas as pd 
import numpy as np

Aviation_data2 = pd.read_csv('AviationData.csv', encoding='latin1')
Aviation_data2.head()
```

# Filling the Missing Values

#### Filling the missing values from backwords using values below the missing value

```python
Aviation_data2.fillna(method='bfill', inplace=True)
Aviation_data2
```

## filling missing values using values above it

```python
Aviation_data2.fillna(method='pad', inplace=True)
Aviation_data2
```

## Print the number of missing values in Dataset

```python
Aviation_data2.isnull().sum()
```

## Drop unnecessary columns in the Dataset

#### This code below allows you to drop columns of the dataset either permanently of temporarily

```python
Aviation_data2.drop(columns=['Event.Id', 'Invest.Type', 'Accident.Number', 'Country', 'Airport.code', 'Airport.Name', 'Schedule', 'Weather.Condition', 'Broad.phase.flight'], inplace=True)
```
